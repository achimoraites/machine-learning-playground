# Bag of words (BOW)

This is a simplistic demonstration how BOW is working.

In a real scenario the vocabulary would be more fine tuned to address the given business problem and the sentence parsing would be more elaborate (handle punctuation etc...)

This is intended to just demonstrate how word vectorization works when using the BOW model 😁 

```python
# Assuming that for our purposes, this vocabulary contains the most important words
vocabulary = ['achilles', 'loves', 'cats', 'but', 'does', 'not', 'like', 'snakes', 'are']
sentence1 = 'Achilles loves cats'
sentence2 = 'Cats are not like snakes'
sentence3 = 'Achilles does not like snakes'


def bow_boolean(sentence, vocab):
    vector = []
    words_in_sentence = set(sentence.lower().split())
    for v_word in vocab:
        if v_word in words_in_sentence:
            vector.append(1)
        else:
            vector.append(0)
    return vector

def bow_count_of_words(sentence, vocab):
    vector = {vocab[i]: 0 for i in range(0, len(vocab))}
    words_in_sentence = sentence.lower().split()
    for word in words_in_sentence:
        if word in vocab:
            vector[word] = vector[word] + 1
    return list(vector.values())


def print_bow_vector(sentence, bow):
    print('The sentence "{}" gets vectorized as {}'.format(sentence, bow(sentence, vocabulary)))
    

print('\n\n### \t BOW Vectorization using boolean representation \n')
print_bow_vector(sentence1, bow_boolean)
print_bow_vector(sentence2, bow_boolean)
print_bow_vector(sentence3, bow_boolean)
print_bow_vector('Cats do not usually like other cats but they get along', bow_boolean)
# It will ingnore any word that is not in the vocabulary!
print_bow_vector('This is a hot pepper be careful', bow_boolean)

print('\n\n### \t BOW Vectorization using Count of Words representation \n')
print_bow_vector(sentence1, bow_count_of_words)
print_bow_vector(sentence2, bow_count_of_words)
print_bow_vector(sentence3, bow_count_of_words)
print_bow_vector('Cats do not usually like other cats but they get along', bow_count_of_words)
# It will ingnore any word that is not in the vocabulary!
print_bow_vector('This is a hot pepper be careful', bow_count_of_words)
```
